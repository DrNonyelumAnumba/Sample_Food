```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
sns.set()
```

```python
data=pd.read_csv(r'foodhub_order.csv')
data
```

```python
y = data ['cost_of_the_order']
x1 = data ['delivery_time']
```

```python
plt.scatter(x1,y)
plt.xlabel('time', fontsize = 20)
plt.ylabel('cost', fontsize = 20)
plt.show()
```

```python
p = sm.add_constant(x1)
results = sm.OLS(y,x1).fit()
results.summary()
```
